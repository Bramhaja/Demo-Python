```python
a = input("Enter a value in the format key:value,key:value :")

print("Your value is:", a)

print("data type is:", type(a))

# Split the string by comma to get key-value pairs
pairs = a.split(',')

# Create an empty dictionary
b = {}

# Iterate through the pairs and add them to the dictionary
for pair in pairs:
    # Split each pair by colon to get key and value
    key_value = pair.split(':')
    if len(key_value) == 2:
        # Convert key to integer and add to dictionary
        b[int(key_value[0])] = key_value[1]
    else:
        print(f"Skipping invalid pair: {pair}")

print("After converting the value is:", b)
print("data type is:", type(b))
```
